```python
#!/usr/bin/python

import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import random as rand
```

```python
X = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])  
m1 = np.mean(X[:, 0])  # 数据中心化：使得均值为零
m2 = np.mean(X[:, 1])
m = np.array([m1, m2])
X = X - m

pca = PCA()
pca.fit(X)
new_axis = pca.components_

print('新座标轴向量,即特征向量：\n', new_axis[0, :], '\n', new_axis[1, :], '\n')
print('对应的特征值:\n', pca.explained_variance_, '\n')  
print('协方差矩阵：\n', np.dot(X.T, X), '\n')
Y = np.dot(new_axis, X.T)  # 新坐标系中的坐标(数据)
print('协方差矩阵：\n', np.dot(Y, Y.T), '\n')

axes = plt.gca()
axes.set_xlim([-5, 5])
axes.set_ylim([-5, 5])
plt.scatter(X[:, 0], X[:, 1], c='gray')
plt.plot([0, 6*new_axis[0, 0]], [0, 6*new_axis[0, 1]], color ='r')  # 新坐标轴1
plt.plot([0, 6*new_axis[1, 0]], [0, 6*new_axis[1, 1]], color = 'b')  # 新坐标轴2
plt.axis('equal')   # 坐标刻度一致
```

```python
X = []
Y = []

NUM = 1000
for i in range(NUM):   # square
    X.append(rand.uniform(-10, 10))
    Y.append(rand.uniform(-10, 10))
m1 = np.mean(X)
m2 = np.mean(Y)
M = np.array([m1, m2]).reshape(2, 1)
D = np.array([X, Y])
D = D - M
D = D.T

pca = PCA()
pca.fit(D)
new_axis = pca.components_

print('新座标轴向量,即特征向量：\n', new_axis[0, :], '\n', new_axis[1, :])
print('对应的特征值:\n', pca.explained_variance_)  
print('协方差矩阵：\n', np.dot(D.T, D), '\n')
D1 = np.dot(new_axis, D.T)  # 新坐标系中的坐标(数据)
print('协方差矩阵：\n', np.dot(D1, D1.T), '\n')

axes = plt.gca()
axes.set_xlim([-5, 5])
axes.set_ylim([-5, 5])

plt.scatter(X, Y, c='gray')

plt.plot([0, 12*new_axis[0, 0]], [0, 12*new_axis[0, 1]], color ='r')  # 新坐标轴1
plt.plot([0, 12*new_axis[1, 0]], [0, 12*new_axis[1, 1]], color = 'b')  # 新坐标轴2
plt.axis('equal')   # 坐标刻度一致
```

```python
X = []
Y = []

NUM = 1000
for i in range(NUM):   # ellipse
    x = rand.uniform(-10, 10)
    y = rand.uniform(-10, 10)
    d1 = np.sqrt((x-6.5)**2 + (y-6.5)**2)
    d2 = np.sqrt((x+6.5)**2 + (y+6.5)**2)
    if d1 + d2 <= 20:
        X.append(x)
        Y.append(y) 

m1 = np.mean(X)
m2 = np.mean(Y)
M = np.array([m1, m2]).reshape(2, 1)       
D = np.array([X, Y])
D = D - M
D = D.T

pca = PCA()
pca.fit(D)
new_axis = pca.components_

print('新座标轴向量,即特征向量：\n', new_axis[0, :], '\n', new_axis[1, :])
print('对应的特征值:\n', pca.explained_variance_)  
print('协方差矩阵：\n', np.dot(D.T, D), '\n')
D1 = np.dot(new_axis, D.T)  # 新坐标系中的坐标(数据)
print('协方差矩阵：\n', np.dot(D1, D1.T), '\n')

axes = plt.gca()
axes.set_xlim([-10, 10])
axes.set_ylim([-10, 10])

plt.scatter(X, Y, c='gray')

plt.plot([0, 12*new_axis[0, 0]], [0, 12*new_axis[0, 1]], color ='r')  # 新坐标轴1
plt.plot([0, 12*new_axis[1, 0]], [0, 12*new_axis[1, 1]], color = 'b')  # 新坐标轴2
plt.axis('equal')   # 坐标刻度一致
```
